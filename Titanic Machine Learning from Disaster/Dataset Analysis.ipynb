```python
import pandas as pd
import math
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head(6)
```

```python
non_null = df[df['Age'].isnull() == False]
```

```python


def fill_missing_age(row):
    if math.isnan(row['Age']):
        sex = row['Sex']
        pclass = row['Pclass']
        return non_null[(non_null['Sex'] == sex) & (non_null['Pclass'] == pclass)]['Age'].median()
    else:
        return row['Age']
                    

df['Age'] = df.apply(fill_missing_age, axis=1)
```
