```python
import pandas as pd
```

```python
df = pd.read_csv('data/stage_1_train.csv')
```

```python
df.head(10)
```

```python
def create_columns(row, subtype):
    if str(row[0]).endswith(subtype):
        return row[1]
    
    return 0

def add_columns_for_subtypes(dataset, subtype):
    df[subtype] = df.apply(lambda row: create_columns(row, subtype), axis=1)

add_columns_for_subtypes(df, 'epidural')
add_columns_for_subtypes(df, 'intraparenchymal')
add_columns_for_subtypes(df, 'intraventricular')
add_columns_for_subtypes(df, 'subarachnoid')
add_columns_for_subtypes(df, 'subdural')
add_columns_for_subtypes(df, 'any')
```

```python
df.head(10)
```

```python
def strip_name_of_subtype(row):
    splitted_name = str(row[0]).split('_')
    return '{0}_{1}'.format(splitted_name[0], splitted_name[1])

df["ID"] = df.apply(lambda row: strip_name_of_subtype(row), axis=1)
```

```python
df.head(10)
```

```python
df.drop(['Label'], axis=1, inplace=True)

aggregation_functions = { 'epidural': 'max', 'intraparenchymal': 'max', 'intraparenchymal': 'max', 'intraventricular': 'max', 'subarachnoid': 'max', 'subdural': 'max', 'any': 'max'}
df_nice = df.groupby(df['ID']).aggregate(aggregation_functions)
df_nice.head(10)
```

```python
df_nice.to_csv('stage_1_train_nice.csv')
```
